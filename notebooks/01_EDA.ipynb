```python
# Step 3: Load dataset and check complete description

import pandas as pd

# Dataset ka path
file_path = "/content/drive/MyDrive/Projects/Machine_Learning/Course_Recomendatioan_System/data/raw/udemy_course_data.csv"

# 1. CSV load
df = pd.read_csv(file_path)

# âœ… Confirmation
print("âœ… Dataset successfully loaded into DataFrame\n")

# 2. Shape of dataset
print(f"ðŸ“Š Dataset Shape -> Rows: {df.shape[0]}, Columns: {df.shape[1]}\n")

# 3. Sample rows
print("ðŸ”¹ Sample Rows (first 5):")
display(df.head())

# 4. Dataset info
print("\nðŸ”¹ Dataset Info:")
df.info()

# 5. Dataset statistics
print("\nðŸ”¹ Dataset Description:")
display(df.describe(include="all"))

# 6. Missing values
print("\nðŸ”¹ Missing Values in each column:")
print(df.isnull().sum())

# 7. Duplicate rows
print(f"\nðŸ”¹ Duplicate Rows: {df.duplicated().sum()}")
```

```python
# Step 4: Data Cleaning & Preprocessing

# Copy dataset for cleaning
df_clean = df.copy()

# 1. Drop missing values
df_clean = df_clean.dropna()

# 2. Remove duplicate rows
df_clean = df_clean.drop_duplicates()

# 3. Convert 'content_duration' to numeric (hours)
# Some values are like '1.5 hours' or '45 mins'
def convert_duration(x):
    try:
        if "hour" in x:
            return float(x.split()[0])  # extract numeric hours
        elif "min" in x:
            return round(float(x.split()[0]) / 60, 2)  # convert minutes to hours
        else:
            return float(x)  # already numeric
    except:
        return None

df_clean["content_duration"] = df_clean["content_duration"].astype(str).apply(convert_duration)

# Drop rows where duration couldn't be parsed
df_clean = df_clean.dropna(subset=["content_duration"])

# 4. Drop unwanted columns
df_clean = df_clean.drop(columns=["url", "published_timestamp"], errors="ignore")

# ============================================================
# âœ… Final check after cleaning
# ============================================================
print("âœ… Shape after cleaning:", df_clean.shape)
print("\nðŸ”¹ Missing values after cleaning:\n", df_clean.isnull().sum())
print("\nðŸ”¹ Dataset Info after cleaning:\n")
print(df_clean.info())
```

```python
import re

# Step X: Text Cleaning for course_title and subject
def clean_text(text):
    # Lowercase
    text = text.lower()

    # Remove special characters, numbers, hyphens etc. but keep letters and spaces
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Remove common stopwords (like and, or, etc.)
    stopwords = ["and", "or", "etc"]
    text = ' '.join([word for word in text.split() if word not in stopwords])

    return text

# Apply cleaning on specific text columns
df['course_title'] = df['course_title'].apply(clean_text)
df['subject'] = df['subject'].apply(clean_text)

print("âœ… Text columns cleaned successfully!")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Ignore all warnings
warnings.filterwarnings("ignore")

# Set seaborn style
sns.set(style="whitegrid")

# 1. Distribution of Course Prices
plt.figure(figsize=(13,6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title("Distribution of Course Prices", fontsize=14)
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
# 2. Number of Courses by Subject
plt.figure(figsize=(13,6))
sns.countplot(data=df, x='subject', order=df['subject'].value_counts().index, palette="Set2")
plt.title("Number of Courses by Subject", fontsize=14)
plt.xlabel("Subject")
plt.ylabel("Number of Courses")
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Average Subscribers per Level
plt.figure(figsize=(13,6))
sns.barplot(data=df, x='level', y='num_subscribers', estimator=lambda x: sum(x)/len(x), palette="muted")
plt.title("Average Subscribers by Course Level", fontsize=14)
plt.xlabel("Course Level")
plt.ylabel("Average Subscribers")
plt.show()
```

```python
# 4. Content Duration Distribution
plt.figure(figsize=(13,6))
sns.histplot(df['content_duration'], bins=30, kde=True, color="orange")
plt.title("Distribution of Content Duration (hours)", fontsize=14)
plt.xlabel("Duration (hours)")
plt.ylabel("Count")
plt.show()
```

```python
# 5. Profit vs Number of Subscribers (scatter plot)
plt.figure(figsize=(13,6))
sns.scatterplot(data=df, x='num_subscribers', y='profit', alpha=0.6)
plt.title("Profit vs Number of Subscribers", fontsize=14)
plt.xlabel("Number of Subscribers")
plt.ylabel("Profit")
plt.show()
```

```python
# 6. Courses Published per Year
plt.figure(figsize=(13,6))
sns.countplot(data=df, x='year', palette="coolwarm")
plt.title("Number of Courses Published per Year", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Number of Courses")
plt.show()
```
